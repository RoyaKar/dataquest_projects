## Data cleaning

### Cleaning Column Names
Converting the column names from camelcase to snakecase and reword some of the column names based on the data dictionary to be more descriptive.

```python
autos.columns
```

```python
import re

new_columns = []

def clean_col(col):
    col = col.replace("yearOfRegistartion","registration_year")
    col = col.replace("monthOfRegistration", "registration_month")
    col = col.replace("yearOfRegistration", "registration_year")
    col = col.replace("notRepairedDamage", "unrepaired_damage")
    col = col.replace("dateCreated", "ad_created")
    col = re.sub(r'(?<!^)(?=[A-Z])', '_', col).lower()
    return col


for names in autos.columns:
    cleaned_col = clean_col(names)
    new_columns.append(cleaned_col)
    
autos.columns = new_columns
```

```python
autos["brand"].value_counts(normalize=True)[:6]
```

Of the top 6 most popular car brands listed, Renault cars are the cheapest cars on average and the most expensive car brand (on average) is Mercedes Benz.  

```python
# Create an empty dictionary to store the results
brand_mean_prices = {}

#Create an array of the top 6 brands
autos_top6 = autos["brand"].value_counts(normalize=True)[:6].index.tolist()

#Use a for loop to iterate over the brands
for auto in autos_top6:
    selected_rows = autos[autos["brand"] == auto]
    mean = selected_rows["price_$"].mean()
    brand_mean_prices[auto] = mean
  

#sorting the dictionary by the lowest price

{k: v for k, v in sorted(
    brand_mean_prices.items(), key=lambda item: item[1]
)}
```

Calculating mean mileage (for the next exercise)

```python
# Create an empty dictionary to store the results
brand_mean_mileage = {}

#Create an array of the top 6 brands
autos_top6 = autos["brand"].value_counts(normalize=True)[:6].index.tolist()

#Use a for loop to iterate over the brands
for auto in autos_top6:
    selected_rows = autos[autos["brand"] == auto]
    mean = selected_rows["odometer_km"].mean()
    brand_mean_mileage[auto] = mean
  

{k: v for k, v in sorted(
    brand_mean_mileage.items(), key=lambda item: item[1]
)}
```

### Average mileage for the top 6 car brands

For the top 6 brands, will use aggregation to understand the average mileage for those cars and if there's any visible link with mean price. 

```python
# Dictionary to series 

bmp_series = pd.Series(brand_mean_prices)
print(bmp_series)

print('\n')

bmm_series = pd.Series(brand_mean_mileage)
print(bmm_series)
```

```python
# Series to Dataframe

brands = pd.DataFrame(bmp_series, columns=['mean_price'])
brands
```

```python
## Assigning bmm to above Dataframe

brands["mean_mileage"] = bmm_series
brands
```

```python
pd.options.display.float_format = '{:,.0f}'.format
brands
```

```python
brands["value_for_money"] =  (brands["mean_price"]) / (brands["mean_mileage"])
brands["value_for_money"] = brands["value_for_money"].map('${:,.3f}'.format)
brands
```

The most cost-effective/value for money brand i.e. lowest price per mileage is Renault.
